# Heston Chebyshev Performance

Read performance results from file and calculate trimmed average.

```python
import numpy as np
import pandas as pd
```

```python
table = pd.read_csv('heston_chebyshev_performance.csv', sep=';', index_col=0)
table
```

```python
res = []
for Nd in table['Nd'].drop_duplicates():
    data = table[table['Nd']==Nd]
    row = {}
    for col in data:
        row[col] = data[col].sort_values()[1:-1].mean()
    res.append(row)
pd.DataFrame(res)
```
